```python
import pandas as pd

overview_df = pd.read_csv('lego_store_overview.csv')

overview_df


```

```python
print(len(overview_df))

# Drop duplicate rows
df2 = overview_df.drop_duplicates(keep='first')

print(len(df2))
df2 = df2[df2['piece_count'] != 'Piece count not found']
print(len(df2))

# Convert piece_count to int
df2['piece_count'] = df2['piece_count'].astype(int)

# Drop rows where piece count is < 10
df2 = df2[df2['piece_count'] >= 10]

print(len(df2))
```

```python

df2['price_per_piece'] = df2['sale_price'].str.replace('$', '').astype(float) / df2['piece_count'].astype(float)
print(len(df2))
df2.sort_values(by='price_per_piece', ascending=True).head(40)
```

```python
df_pov = pd.read_csv('pov.csv')
df_pov['pov_past_6_months'] = df_pov['pov_past_6_months'].str.replace('US $', '')
df_pov['pov_past_6_months'] = df_pov['pov_past_6_months'].str.replace(',', '')
df_pov['pov_past_6_months'] = pd.to_numeric(df_pov['pov_past_6_months'])

df_pov['pov_current_listings'] = df_pov['pov_current_listings'].str.replace('US $', '')
df_pov['pov_current_listings'] = df_pov['pov_current_listings'].str.replace(',', '')
df_pov['pov_current_listings'] = pd.to_numeric(df_pov['pov_current_listings'])

# Get raw profit
df_pov['pov_past_6_months_profit'] = df_pov['pov_past_6_months'].astype(float) - df_pov['sale_price'].str.replace('$', '').astype(float)
df_pov['pov_current_listings_profit'] = df_pov['pov_current_listings'].astype(float) - df_pov['sale_price'].str.replace('$', '').astype(float)

# Get profit as a percentage of sale price
df_pov['pov_past_6_months_profit/sale_price'] = df_pov['pov_past_6_months_profit'] / df_pov['sale_price'].str.replace('$', '').astype(float)
df_pov['pov_current_listings_profit/sale_price'] = df_pov['pov_current_listings_profit'] / df_pov['sale_price'].str.replace('$', '').astype(float)  

df_pov.sort_values(by='pov_past_6_months_profit/sale_price', ascending=False).tail(40)
```
